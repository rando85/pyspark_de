```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()
```

```python
df = spark.read.parquet('./data/dataframe')
```

```python
df.printSchema()
```

```python
df.show()
```

```python
#count
from pyspark.sql.functions import count

df.select(
    count('nombre').alias('conteo_nombre'),
    count('color').alias('conteo_colores'),
    count('*').alias('conteo_general')
).show()
```

```python
#countDistinct
from pyspark.sql.functions import countDistinct

df.select(
    countDistinct('color').alias('colores_diferentes')
).show()
```

```python
#approx_count_distinct
from pyspark.sql.functions import approx_count_distinct

dataframe = spark.read.option("header", "true").option('inferSchema', "true").csv('./data/flights.csv')
```

```python
dataframe.printSchema()
```

```python
dataframe.select(
    countDistinct('AIRLINE'),
    approx_count_distinct('AIRLINE'),
).show()
```
