```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()
```

```python
estudiantes = spark.read.parquet('./data/estudiantes.parquet')
```

```python
estudiantes.show()
```

```python
from pyspark.sql.functions import min, max, avg, col
```

```python
estudiantes.groupBy('graduacion').pivot('sexo').agg(avg('peso')).show()
```

```python
estudiantes.groupBy('graduacion').pivot('sexo').agg(
    avg('peso'),
    min('peso'),
    max('peso')
).show()
```

```python
estudiantes.groupBy('graduacion').pivot('sexo', ['M']).agg(
    avg('peso'),
    min('peso'),
    max('peso')
).show()
```

```python
estudiantes.groupBy('graduacion').pivot('sexo', ['F']).agg(
    avg('peso'),
    min('peso'),
    max('peso')
).show()
```
